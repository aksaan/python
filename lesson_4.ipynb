```python
import matplotlib.pyplot as plt
x = list(range(-100, 101))
y = list(map(lambda y: y ** 2, x))
plt.plot(x, y)
```

```python
x = list(range(-100, 101))
x.remove(0)
y = list(map(lambda y: 1 / y, x))
plt.scatter(x, y, color = "red")
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.title("гипербола")
plt.xlabel("x")
plt.ylabel("y")
```

```python
df = pd.read_csv('Store.csv')
states = df["State"].unique()
states
```

```python
sales = []
for state in states:
  sales.append(df.loc[df["State"] == state, "Sales"].sum())
sales
```

```python
plt.figure(figsize = (100, 40))
plt.bar(states, sales, label = "Сумма продаж", )
plt.title("Продажи")
```

```python
categories = df["Category"].unique()
categories
```

```python
income = []
for category in categories:
  income.append(df.loc[df["Category"] == category, "Profit"].mean())
income
```

```python
plt.figure(figsize = (40, 20))
plt.bar(categories, income, label = "Доход")
plt.legend()
# вывод label
```
