```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('Store.csv')
df.head()
```

```python
categories = df["Category"].unique()
categories
```

```python
incomes = []
min_ground = -500
max_ground = 4000
for category in categories:
  incomes.append(df.loc[(df["Category"] == category) & (df["Profit"] >= min_ground) & (df["Profit"] <= max_ground), "Profit"])
```

```python
plt.boxplot(incomes, patch_artist=True, autorange=True, vert=0)
```

```python
np.random.seed(100)

d = np.random.normal(150, 10, 250)
d1 = np.random.normal(140, 20, 250)
d2 = np.random.normal(130, 30, 250)
data = [d, d1, d2]

plt.boxplot(data, patch_artist=True, vert=0)
plt.show()
print(d, d1, d2)
```
