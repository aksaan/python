Построить столбиковую диаграмму, которая показывает распределение:
сколько покемонов с силой

*  менее 50
*  от 50 до 100
*  более 100

```python
categories = [
    len(df[df["Attack"] < 50]),
    len(df[(df["Attack"] >= 50) & (df["Attack"] < 100)]),
    len(df[df["Attack"] >= 100])
]
plt.bar(height = categories, x = ["Менее 50", "От 50 до 100", "Более 100"])
```
